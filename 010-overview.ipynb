# Human vs OCR

Human Reading:
- sequential, variable resolution
- fixations, tracking text lines
- highly robust

OCR (even with DL):
- highly parallel, single resolution
- parallel recognition
- sensitive to deviations from training data

# Advantages of Human Reading Technique

- highly robust to page curl, distortions, etc.
- dynamic resources allocation
- integrates layout, reading, reading order, language modeling
- self-teaching: scales from character-by-character to multi-word
- backtracking for corrections and self-teaching

# Human vs Machine Reading

The differences between human reading and OCR are driven by:

- computational hardware
- available DL / ML / image processing techniques

Experimental human-like OCR is interesting, but engineered solutions (even DL based) will continue to work differently for many years.

# Solved vs Open

Solved:

- OCR of flatbed scanned, printed texts achieves error rates of < 0.3% with LSTM

Open (at least to some degree):

- analysis and understanding of complex layouts
- reliable reading order of complex / multi-column documents
- table understanding
- figure understanding
- high quality reconstruction of camera captured documents
- high quality OCR from camera captured documents


# Covered Today

# Specific Techniques Covered Today

- dlinputs - large scale, distributed storage and input pipelines
- ocrodeg - document image degradation for data augmentation
- ocrorot - page rotation and skew detection
- ocrobin - document image binarization using 2D LSTM
- ocroseg - page layout analysis using 2D LSTM
- ocroline - text line recognition using 1D LSTM and CTC
- ocrolm - language modeling using 1D LSTM



# Not Covered (But Worthwhile)

- purely convolutional models
- seq2seq models from speech (might work for OCR, OCR language modeling)
- word-based language models
- document image dewarping
- scene text recognition
- reinforcement learning and similar techniques for human-like OCR
- completely unsupervised training
